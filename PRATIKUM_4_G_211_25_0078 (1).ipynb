```python
def linear_search(arr, target):
    # Menggunakan enumerate untuk iterasi indeks dan elemen sekaligus
    for idx, element in enumerate(arr):  #cek apakah elemen sama dengan target
        if element == target:            # jika ketemu, kembalikan indeks
            return idx                    # jika tidak ditemukan
    return -1

data = [10, 20, 30, 40, 50]
print(linear_search(data, 30))  # Output: 2
```

```python
def binary_search(arr, target):
    rendah, tinggi = 0, len(arr) - 1

    while rendah <= tinggi:
        tengah = rendah + (tinggi - rendah) // 2    # batas kiri dan kanan array
        if arr[tengah] == target:            # selama masih ada ruang pencarian
            return tengah                    # cari indeks tengah return tengah
        elif arr[tengah] < target:                 # jika nilai tengah = target
            rendah = tengah + 1                    # kembalikan indeks
        else:                                      # jika target lebi
            tinggi = tengah - 1                    # geser ke kanan
                                                   # jika target lebih kecil
                                                      # geser ke kiri
    return -1                                         # jika tidak ditemukan

data = [10, 20, 30, 40, 50]
print(binary_search(data, 40))  # Output: 3
```

```python
def binary_search_recursive(arr, mulai, akhir, target):
    if mulai > akhir:
        return -1

    titik_tengah = (mulai + akhir) // 2

    if arr[titik_tengah] == target:
        return titik_tengah
    elif arr[titik_tengah] < target:
        return binary_search_recursive(arr, titik_tengah + 1, akhir, target)
    else:
        return binary_search_recursive(arr, mulai, titik_tengah - 1, target)

data = [10, 20, 30, 40, 50]
print(binary_search_recursive(data, 0, len(data)-1, 20))  # Output: 1
```

```python
import math

def jump_search(arr, target):
    n = len(arr)
    lompatan = int(math.sqrt(n))
    sebelumnya = 0

    # Mencari blok di mana target mungkin berada
    while arr[min(lompatan, n) - 1] < target:
        sebelumnya = lompatan
        lompatan += int(math.sqrt(n))
        if sebelumnya >= n:
            return -1

    # Melakukan linear search pada blok yang ditemukan
    for i in range(sebelumnya, min(lompatan, n)):
        if arr[i] == target:
            return i

    return -1

data = [10, 20, 30, 40, 50, 60, 70]
print(jump_search(data, 60))  # Output: 5
```

```python
def interpolation_search(arr, target):
    low, high = 0, len(arr) - 1

    while low <= high and arr[low] <= target <= arr[high]:
        if low == high:
            return low if arr[low] == target else -1

        # Rumus posisi interpolasi
        pos = low + int(((target - arr[low]) * (high - low)) / (arr[high] - arr[low]))

        if arr[pos] == target:
            return pos
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1

    return -1

data = [10, 20, 30, 40, 50]
print(f"Index of 30: {interpolation_search(data, 30)}") # Output: 2
print(f"Index of 50: {interpolation_search(data, 50)}") # Output: 4
print(f"Index of 15: {interpolation_search(data, 15)}") # Output: -1
```
