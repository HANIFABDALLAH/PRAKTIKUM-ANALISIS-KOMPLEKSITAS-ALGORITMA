```python
ALGORITMA INSERT

Input : key
Output : data masuk ke tree

Buat node baru

Jika root kosong maka
    root = node baru
    Selesai

current = root

Selama TRUE lakukan

    Jika key < current.key maka

        Jika current.left kosong maka
            current.left = node baru
            Keluar dari perulangan
        Jika tidak
            current = current.left

    Jika tidak

        Jika current.right kosong maka
            current.right = node baru
            Keluar dari perulangan
        Jika tidak
            current = current.right

Tampilkan "Data berhasil ditambahkan"
```

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None


class RedBlackTree:
    def __init__(self):
        self.root = None

    def insert(self, key):
        node = Node(key)

        if self.root is None:
            self.root = node
            print("Data", key, "berhasil ditambahkan")
            return

        current = self.root

        while True:
            if key < current.key:
                if current.left is None:
                    current.left = node
                    break
                current = current.left
            else:
                if current.right is None:
                    current.right = node
                    break
                current = current.right

        print("Data", key, "berhasil ditambahkan")

    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.key, end=" ")
            self.inorder(node.right)


# Program utama
tree = RedBlackTree()

tree.insert(10)
tree.insert(20)
tree.insert(30)
tree.insert(15)
tree.insert(25)

print("\nData dalam tree:")
tree.inorder(tree.root)
```
